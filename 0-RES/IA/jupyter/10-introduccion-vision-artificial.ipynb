```python
# importar paquetes

from matplotlib import pyplot as plt
import numpy as np
import imageio.v2 as imageio
from skimage import color
```

```python
# Lectura de la imagen

ruta_archivo = "./public/img/jpg/gray-wolf.jpg"

imgInput = imageio.imread(ruta_archivo)
```

```python
print("Dimensaiones de la imagen ", imgInput.shape )
```

```python
# imprimir el contenido de las capas de color

print("Pixel en [0,0,0] ", imgInput[0,0,0])
print("Pixel en [0,0,1] ", imgInput[0,0,1])
print("Pixel en [0,0,2] ", imgInput[0,0,2])
print("Pixel en [0,0,:] ", imgInput[0,0,:])
```

```python
print("Tipo de dato: ", imgInput.dtype)
```

```python
# rango de valores contenidos en la matriz (por color)
print('Valor maximo ', imgInput.max())
print('Valor minimo ', imgInput.min())
```

# Visualizacion de la imagen

```python
plt.imshow(imgInput);
```

# Transformaciones


## escala de grises

```python
plt.imshow(imgInput[:,:,0], cmap='gray');
```

```python
imgGray = color.rgb2gray(imgInput)
print("Dimensaiones de la imagen ", imgGray.shape )
print("Tipo de dato: ", imgGray.dtype)
print('Valor maximo ', imgGray.max())
print('Valor minimo ', imgGray.min())
```

```python
# imagen en escala de grises? veamos el error

plt.imshow(imgGray);
```

```python
# corregido, ya no lo trata como RGB truncando los valores
plt.imshow(imgGray, cmap='gray');
```

```python
plt.imshow(np.transpose(imgInput, (1, 0, 2)));
```

## SECCION DE LA IMAGEN

```python
imgSeccion = imgGray[100:440, 215:525]
plt.imshow(imgSeccion, cmap='gray');
```

### Implementar filtros

```python
# cambio de colores en la matriz de color
imgModificada = imgGray.copy()
imgModificada[imgModificada < 0.2] = 0.0
plt.imshow(imgModificada, cmap='gray');
```

```python
# distribucion de los valores de la matriz
plt.hist(imgGray.flatten(), bins=64, range=[0,1]);
```

```python
plt.hist(imgGray.flatten());
```

```python
#   histograma de la imagen                de 0 a 1 dividido en 256 porciones
plt.hist(imgGray.flatten(), bins=np.linspace(0,1,256));
```

```python
imageio.imwrite('./public/img/jpg/lobo-modificada2.jpg', (255*imgModificada).astype(np.uint8))
```
