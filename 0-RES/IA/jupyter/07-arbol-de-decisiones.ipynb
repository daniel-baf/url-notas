```python
# BIBLIOTECAS BASICAS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# CLASIFICADOR, METRICAS Y SEPARADOR DE DATOS DE PRUEBA

# importar el separador de muestras para entrenamiento y pruebas
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
# importar las metricas de evaluacion
from sklearn import metrics
```

```python
# CARGAR DATASET
data = sns.load_dataset('iris')
```

```python
# Separacion de datos en Entrenamiento y Prueba
train_data, test_data = train_test_split(data, test_size=0.4, stratify=data['species'], random_state=42)
```

```python
# separacion de la etiqueta de clase (variable endogena) «Y» y las variables independientes (exogenas o explicativas) «X»
# Conjunto de entrenamiento
x_train = train_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_train = train_data['species']

# Conjunto de pruebas
x_test = test_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_test = test_data['species']
```

```python
# Arbol de decision
# dt = decission tree
model_dt = DecisionTreeClassifier(random_state=1, max_depth=3)
# ajustar / entrenar el modelo
model_dt.fit(x_train, y_train)


# Predecir resultados sobre el conjunto de pruebas
prediction = model_dt.predict(x_test)
```

```python
# Importancia de cada predictor
# la importancia es que tanto influye un atributo en la toma de decisiones del modelo
model_dt.feature_importances_
```

```python
# Etiquetas de datos
factores = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] # parametros a analizar
categories = ["setosa", "versicolor", "virginica"] # especies a analizar, segun el orden en que se encontro en el diccionario
```

```python
# Graficar el arbol de decision
plt.figure(figsize=(18, 15))
plot_tree(model_dt, filled=True, feature_names=factores, class_names=categories);
```

```python
# Eficiencia
accuracy = metrics.accuracy_score(y_test, prediction)
accuracy
```
