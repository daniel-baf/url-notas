**Ejemplo de grafica en una presentacion interactiva**

$$\text{Sea } 0 \leqslant x \leqslant 10$$
$$f_1(x) = x$$
$$f_2(x) = x^2$$
$$f_3(x) = x^3$$

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Rango de valores para x
#   np.arrage(min, max, paso)
#           inicia en min, y termina en max - 1
x = np.arange(0, 10, 0.4)
x
````

````python
# graficar una funcion lineal
# plot(x,y, formato)

plt.plot(x, x, 'g--');
````

````python
y1 = x
y2 = x**2 # x^2
y3 = x**3 # x^3
````

````python
# graficar 3 funciones en un mismo marco de graficos
# s = square, -- = dot
plt.plot(x, y1, 'g--', x, y2, 'bs', x, y3, 'r^')
plt.show()
````

````python
# Agrupaciones para conocer por sexo

print(data['sex'].value_counts())
````

````python
print(data["species"].value_counts())
````

````python
# Group by
data.groupby(by=["species", "sex"]).describe().T
````

````python
data.groupby(by=["species", "sex"]).mean(numeric_only=True)
````

````python
# cols a extraer
numeric_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

# group by  + filter
data.groupby(by=["species", "sex"])[numeric_cols].aggregate(
    ["min", np.median, max, np.mean]
).T
````
