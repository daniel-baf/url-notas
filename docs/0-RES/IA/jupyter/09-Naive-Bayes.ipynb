# Naive Bayes

es un algoritmo de aprendizaje automatico basado en teorema de bayes, que es una forma de calcular la probabilidad de un evento basandoes en el conocimiento previo de condiciones de ocurrencia de algo.

Es util para problemas de clasificacion. (ej. spam, o filtrar tipo de correo)

$$ P_{(A|B)} = \frac{P_{(B|A)}*P_{A}}{P_{B}} $$

En el contexto de clasificacion, queremos encontrar la probabilidad de que una instancia pertenezca a una clase particular dado un conjunto de caracteristicas.

$$P_{C|F_1,F_2...F_n}=\frac{P(F_1,F_2...F_n|C)*P_c}{P(F_1,F_2...F_n)}$$

Donde C es la clase, y $C|F_1,F_2...F_n$ son las caracteristicas de las instancias

Naive significa "ingenuo", todas las independencias entre si, son independientes, son excluyentes.

- Sexo
- Region
- Endemico
- Edad

... todo esto le llamamos "supuesto", la presencia o ausencia de las caracteristicas no esta relacionado con la otra, porque son excluyentes.
`no por ser mujer, va ser exomorfo`... no hay relaciones causales.

## Pasos para clasificar

>   1. Calcula la probabilidad $P_c$
>   2. Calcula la probabiilidad condicional $P_{F_i|C}$
>   3. Calcula la probabilidad posterior $P_{C|F_1,F_2...F_n}$
>   4. Seleccinar la clase con la mayoro probabilidad posterior

```python
# Bibliotecas basicas

# BIBLIOTECAS BASICAS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# CLASIFICADOR, METRICAS Y SEPARADOR DE DATOS DE PRUEBA

# importar el separador de muestras para entrenamiento y pruebas
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
# importar las metricas de evaluacion
from sklearn import metrics
```

```python
# CARGAR DATASET
data = sns.load_dataset('iris')

# Separacion de datos en Entrenamiento y Prueba
train_data, test_data = train_test_split(data, test_size=0.4, stratify=data['species'], random_state=19)
```

```python
# separacion de la etiqueta de clase (variable endogena) «Y» y las variables independientes (exogenas o explicativas) «X»
# Conjunto de entrenamiento
x_train = train_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_train = train_data['species']

# Conjunto de pruebas
x_test = test_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_test = test_data['species']
```

```python
# clasificador gaussiano Naive Bayes
model_gnb = GaussianNB()
y_prediction = model_gnb.fit(x_train, y_train).predict(x_test)
print("Prediccion del clasificador gaussiano Naive Bayes en los datos de prueba {:0.3f}: ".format(metrics.accuracy_score(y_prediction, y_test)))
```

```python
model_gnb_2_predictors = GaussianNB()
predictors = ['petal_length', 'petal_width']
y_prediction2 = model_gnb_2_predictors.fit(x_train[predictors], y_train).predict(x_test[predictors])
print("Prediccion del clasificador gaussiano Naive Bayes en los datos de prueba con dos predictores "
        "{:0.3f}: ".format(metrics.accuracy_score(y_prediction2, y_test)))
```

```python
test_data["prediccion"] = y_prediction
test_data["prediccion2"] = y_prediction2
test_data
```
