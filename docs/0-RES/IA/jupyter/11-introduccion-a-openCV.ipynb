# Introducción a OpenCV

```python
# importar paquetes
from matplotlib import pyplot as plt
import numpy as np
# import imageio.v2 as imageio
# from skimage import color

# Biblioteca Open CV
import cv2
# import common

# Para control de dimensiones de la imagen
import pylab
```

```python
# modificar las dimensiones de visualización
pylab.rcParams['figure.figsize'] = (6.0, 4.0)

# cargar la imagen
ruta = './public/img/jpg/wall-e.jpg'
input_image = cv2.imread(ruta);
```

```python
plt.imshow(input_image);
# plt.show(input_image);
```

```python
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB));
```

```python
# inversion de la imgaen
flipped_img = cv2.flip(input_image, 0)
plt.imshow(flipped_img);
```

```python
# inversion de la imgaen
flipped_img = cv2.flip(input_image, 1)
plt.imshow(flipped_img);
```

```python
# imagen transpuesta
transpuesta = cv2.transpose(input_image)
plt.imshow(transpuesta);
```

```python
plt.imshow(cv2.cvtColor(cv2.flip(transpuesta, 1), cv2.COLOR_BGR2RGB));
```

# Operaciones aritmeticas

```python
# crear una matriz de h,w,3 con ceros
blank_image = np.zeros((input_image.shape), np.uint8)

# crear el marco
blank_image[105:230, 320:530, 1] = 50
plt.imshow(blank_image);
```

```python
plt.imshow(blank_image);
```

```python
nueva_imagen = cv2.cvtColor(cv2.add(blank_image, input_image), cv2.COLOR_BGR2RGB)
plt.imshow(nueva_imagen);
```

```python
# filtro de desenfoque
desenfoque = 6
img_desenfocada = cv2.GaussianBlur(input_image, (2*desenfoque+1, 2*desenfoque+1), 0)
plt.imshow(cv2.cvtColor(img_desenfocada, cv2.COLOR_BGR2RGB));
```

```python
imagen_gris = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
plt.imshow(imagen_gris, cmap='gray');
```

```python
# FILTR DE SObEL, HACER QUE LOS BORDES SEAN MAS VISIBLES

# Sobel, en este caso, es mas conveniente usar imagenes en escala de grises
sobel_x = cv2.Sobel(imagen_gris, cv2.CV_64F, 1, 0, ksize=9)  # se usa CV_64F porque la escala de grises convierte todo a flotante de 64 bits
# el 0, 1
plt.imshow(sobel_x, cmap='gray');
```

```python
sobel_y = cv2.Sobel(imagen_gris, cv2.CV_64F, 0, 1, ksize=9)
plt.imshow(sobel_y, cmap='gray');
```

```python
sobel_xy = cv2.Sobel(imagen_gris, cv2.CV_64F, 1, 1, ksize=9)
plt.imshow(sobel_xy, cmap='gray');
```

## Maximos y minimos

```python
for i in range (0,3):
    minimo, maximo, min_location, max_location = cv2.minMaxLoc(input_image[:,:,i]) # :,:,i -> todos los pixeles de x, de y, en cada capa
    print(f'Canal {i}: minimo: {minimo}, maximo: {maximo}, min_location: {min_location}, max_location: {max_location}')
```
